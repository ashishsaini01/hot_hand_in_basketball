```python
# paul_df = pd.read_csv('paul_season_games_data.csv')
paul_vs_dallas_df = pd.read_csv('paul_series_games_data.csv')
paul_vs_orleans_df = pd.read_csv('paul_series_games_data(suns_vs_peli).csv')

paul_vs_orleans_df['oponent_team'] = 'orleans'
paul_vs_dallas_df['oponent_team'] = 'dallas'

combine_df = pd.concat([paul_vs_dallas_df, paul_vs_orleans_df])
combine_df['misses/count_binary'] = np.where(combine_df['misses/count'] == 'makes', 1, 0)

combine_df = combine_df[combine_df['basket_type'] != 'free']
```

```python
# paul_vs_orleans_df.shape, paul_vs_dallas_df.shape
combine_df.shape
```

```python
# calculating the streaks 
combine_df['streak'] = combine_df['misses/count_binary'].groupby((combine_df['misses/count_binary'] != combine_df.groupby(['Series Game', 'oponent_team'])['misses/count_binary'].shift()).cumsum()).cumcount() + 1
```

```python
combine_df[['Series Game', 'quarter', 'oponent_team', 'misses/count', 'streak']].tail(20)
```

```python
cross_df = pd.crosstab(combine_df['misses/count'], combine_df['Series Game'])
cross_df = cross_df.reset_index()
cross_df['Total'] = cross_df.sum(axis = 1)
cross_df
```

```python
cross_streak_df = pd.crosstab(combine_df['misses/count'], combine_df['streak'])
cross_streak_df = cross_streak_df.reset_index()
cross_streak_df['Total'] = cross_streak_df.sum(axis = 1)
cross_streak_df
```

```python
makes_prob = cross_streak_df[[1,2,3,4]].loc[0] / 122
misses_prob = cross_streak_df[[1,2,3,4]].loc[1] / 72

prob_df = pd.DataFrame()
prob_df['makes_prob'] = makes_prob
prob_df['misses_prob'] = misses_prob
prob_df = prob_df.reset_index()

prob_df['converg_prob'] = [0.368, 0.563, 0.5, 0.778]
```

```python
prob_df
```

```python
sns.set_style('darkgrid')
ax = sns.barplot(x = 'streak', y = 'converg_prob', data = prob_df, color = 'y')
ax = sns.pointplot(x = 'streak', y = 'converg_prob', data = prob_df, markers = "*")#prob_df[prob_df['streak'] <= 4])
ax.set(xlabel='Short Streaks', ylabel='Probabilitis')
ax.set_title('Hot hand (shorts convergence probability) of chris Paul in Playoffs(2021-22 season)')
```
