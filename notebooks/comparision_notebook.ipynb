```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
curry_df1 = pd.read_csv('curry/curry_series_games(gsw_vs_denver)_data.csv')
curry_df2 = pd.read_csv('curry/curry_series_games(gsw_vs_mem)_data.csv')

print(curry_df1.shape, curry_df2.shape)

# curry_df1['misses/count_binary'] = np.where(curry_df1['misses/count'] == 'makes', 1, 0)
curry_df1 = curry_df1[curry_df1['basket_type'] == '3-pt']
curry_df1 = curry_df1.reset_index(drop = True).reset_index()

# curry_df2['misses/count_binary'] = np.where(curry_df1['misses/count'] == 'makes', 1, 0)
curry_df2 = curry_df2[curry_df2['basket_type'] == '3-pt']
curry_df2 = curry_df2.reset_index(drop = True).reset_index()

curry_df1['misses/count'] = np.where(curry_df1['misses/count'] == 'misses', 'Fehlwürfe', 'Treffer')
curry_df2['misses/count'] = np.where(curry_df2['misses/count'] == 'misses', 'Fehlwürfe', 'Treffer')

print(curry_df1.shape, curry_df2.shape)
```
