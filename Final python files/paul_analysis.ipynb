```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading pre-processed csv files
paul_vs_dallas_df = pd.read_csv('Data/preprocessed_data/paul_data/paul_series_games_data.csv')
paul_vs_orleans_df = pd.read_csv('Data/preprocessed_data/paul_data/paul_series_games_data(suns_vs_peli).csv')

# creating column oponent team
paul_vs_orleans_df['oponent_team'] = 'orleans' 
paul_vs_dallas_df['oponent_team'] = 'dallas'

# filtering data, only for 4th quarter
paul_vs_dallas_df_4th = paul_vs_dallas_df[paul_vs_dallas_df['quarter'] == 4]
paul_vs_orleans_df_4th = paul_vs_orleans_df[paul_vs_orleans_df['quarter'] == 4]

# removing entries of free throws (considering 2-pts and 3-pts)
paul_vs_dallas_df_4th = paul_vs_dallas_df_4th[paul_vs_dallas_df_4th['short_type'] != 'free']
paul_vs_orleans_df_4th = paul_vs_orleans_df_4th[paul_vs_orleans_df_4th['short_type'] != 'free']

# reset index of dataframes
paul_vs_dallas_df_4th.reset_index(drop = True).reset_index(inplace = True)
paul_vs_orleans_df_4th.reset_index(drop = True).reset_index(inplace = True)

# modifiying index column, now it's starts from 1
paul_vs_dallas_df_4th['index'] = [i for i in range(1, paul_vs_dallas_df_4th.shape[0] + 1)]
paul_vs_orleans_df_4th['index'] = [i for i in range(1, paul_vs_orleans_df_4th.shape[0] + 1)]

# changing names to german misses: Fehlwürfe and makes: Treffer
paul_vs_dallas_df_4th['misses/count'] = np.where(paul_vs_dallas_df_4th['misses/count'] == 'misses', 'Fehlwürfe', 'Treffer')
paul_vs_orleans_df_4th['misses/count'] = np.where(paul_vs_orleans_df_4th['misses/count'] == 'misses', 'Fehlwürfe', 'Treffer')
```

```python
# ploting data using seaborn's scatterplots 
# creating two subplots
fig, (ax1, ax2) = plt.subplots(figsize=(11, 5), ncols=2, sharey=True)

# adding text in 1st graph
ax1.text(12, 0.95, 'Win', fontsize = 10)
ax1.text(20, 1.95, 'Loss', fontsize = 10)
ax1.text(35, 2.95, 'Win', fontsize = 10)
ax1.text(37, 3.95, 'Loss', fontsize = 10)
ax1.text(31, 4.95, 'Win', fontsize = 10)
ax1.text(37, 5.95, 'Win', fontsize = 10)

# plotting between index and Series Games
sns.scatterplot(ax = ax1, x = "index", y = "Series Game", hue = "misses/count", alpha = .9, palette = ['g', 'r'],
             data = paul_vs_orleans_df_4th)

ax1.set_xlabel('Anzahl der Würfe') # setting xlabel
ax1.set_ylabel('Spiel') # setting ylabel
ax1.set_title('Playoff-Runde gegen New Orleans') # titles
ax1.get_legend().remove() # removing one legend because both graphs has same legends.


# adding text in 2nd graph
ax2.text(10, 0.95, 'Win', fontsize = 10)
ax2.text(18, 1.95, 'Win', fontsize = 10)
ax2.text(21, 2.95, 'Loss', fontsize = 10)
ax2.text(10, 3.95, 'Not Played (Loss)', fontsize = 10)
ax2.text(24, 4.95, 'Win', fontsize = 10)
ax2.text(20, 5.95, 'Loss', fontsize = 10)
ax2.text(23, 6.95, 'Loss', fontsize = 10)

# plotting
sns.scatterplot(ax = ax2, x = "index", y = "Series Game", hue = "misses/count", alpha = .9, palette = ['r', 'g'],
             data = paul_vs_dallas_df_4th)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5)) # location for legend
ax2.set_xlabel('Anzahl der Würfe') # setting xlabel
ax2.set_title('Playoff-Runde gegen Dallas') # setting title

fig.suptitle('Paul im 4. Viertel in den Playoffs 2022', fontsize=15) # for main title

# plt.savefig('chris_paul.png', bbox_inches='tight', dpi = 120) # saving the graph
```
